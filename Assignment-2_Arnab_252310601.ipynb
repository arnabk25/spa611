**Q.1**

```python
import numpy as np
import matplotlib.pyplot as plt

sigma=5
N=50000  #number of trials

#define voltage
v=np.random.normal(0,sigma,N)

#average power
avg_pow=np.mean(v**2)
variance=np.var(v)

print("Average_power: ",avg_pow)
print("Variance: ",variance)
```

We can see that average (mean) power is almost equal to variance.

**Q.2**

```python
#RMS values
rms_val=[0.5,1,1.5]

N=50000 #trials

#figure plot
plt.figure() 

for rms in rms_val:
  v=np.random.normal(0,rms,N)
  plt.hist(v,bins=100,density=True,alpha=0.5,label=f'RMS={rms}')

plt.legend()

#heading
plt.title("Histograms for RMS Values")

#labeling
plt.xlabel('Voltage')
plt.ylabel('Probability density')

plt.show()
```

**Q.3**

```python
# trials
N=500000

t=np.linspace(0,15*np.pi,N)

#define noise
nois=np.random.normal(0,2,N)

#sine fn
sine=5*np.sin(0.1*t)

#new voltage
v_new=nois+sine

#figure
plt.figure()

#histogram for noise only
plt.hist(nois, bins=100, density=True,alpha=0.5, label="Noise")

#historgram with added sine
plt.hist(v_new, bins=100, density=True,alpha=0.5, label="Noise with sin")

plt.title("Changes for adding slowly varying Sine Wave")

plt.xlabel('Voltage')
plt.ylabel('Probability density')

plt.legend()
plt.show()
```

Histogram is no longer behaving as purely Gaussian. The plot becomes broader and the distribution is more spreaded.

**Q.4**

```python
#ensemble
ens=10000

#true sigma
t_sig=3

#ensemble sample
group=100

vr=[] #variance

for _ in range(ens):
  v=np.random.normal(0,t_sig,group)
  vr.append(np.var(v))
  
plt.figure()

#histogram
plt.hist(vr,bins=100,density=True)

plt.title("Distribution of Sample Variances")

plt.xlabel("Sample Variance")
plt.ylabel("Probability")

plt.show()

print("Average estimated variances:",np.mean(vr))

print("Original variance:", t_sig**2)
```

Graph is skewed. For large values of N, the graph is Gaussian almost.

 Average estimated variances is almost same to the value of original variance.
